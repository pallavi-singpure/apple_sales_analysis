Top 10 highest rated iphone on flipkart?

```python
high_rated=df.sort_values(by=['Star Rating'],ascending=False)
high_rated.head(10)
print(high_rated['Product Name'])
```

How many rating do the highest_rated iphoneson flipkart have

```python
iphone = high_rated['Product Name'].value_counts() # Changed value_count() to value_counts()
label = iphone.index
counts = high_rated['Number Of Ratings'] # Make sure highest_rated is the correct DataFrame name
                                      # and that it contains the column "Number of ratings".
                                      # Maybe you need: counts = high_rated['Number of Ratings']
                                      # or if you meant to use the iphone variable: counts = iph
fig=px.bar(high_rated,x=label, y=counts)
fig.show()
```

which iphone has the highest number of review on flipkart

```python
iphone = high_rated['Product Name'].value_counts() # Changed value_count() to value_counts()
lable=iphone.index
counts=high_rated['Number Of Reviews']
fig=px.bar(high_rated,x=lable,y=counts)
fig.show()
```

what is the relationship between the sales prices of iphone and the number of rating flipkart?

```python
fig=px.scatter(df,x='Sale Price',y='Number Of Ratings',color='Number Of Ratings',title='relationship between the sales prices of iphone and the number of rating',trendline='ols')
fig.show()
```

what is the relationship between the discount precentage and the number of rating of iphone flipkart?

```python
fig=px.scatter(df,x='Discount Percentage',y='Number Of Ratings',color='Number Of Ratings',title='relationship between the discount precentage and the number of rating of iphone flipkart',trendline='ols')
fig.show()
```

most expensive and least expensive iphone
